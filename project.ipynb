# Total Earning in Each Year

This raw data contains earning information for top 500 players every year from 2001-2023. 

This raw data contains in a dictionary.

```python
df_total_earning = {}
# year to scrape
years = [2023,2022,2021,2020,2019,2018,2017,2016,2015,2014,2013,2012,2011,2010,2009,2008,2007,2006,2005,2004,2003,2002,2001]
#list for top100-top500
tops = [100,200,300,400]
for year in years:
    # URL of the page to scrape
    url = f'https://www.esportsearnings.com/history/{year}/top_players'

    # Send a GET request to the URL
    response = requests.get(url)

    # Check if the request was successful
    if response.status_code == 200:
        # Parse the content of the request with BeautifulSoup
        soup = BeautifulSoup(response.text, 'html.parser')
        # Find all divs with the class 'detail_list_box'
        detail_list_boxes = soup.find_all('div', class_='detail_list_box')
        item = detail_list_boxes[1].get_text(separator=' ', strip=True)
    else:
        print(f"Failed to retrieve data. Status code: {response.status_code}")

    item = item.split("% of Total ")[-1]

    # Define a regular expression pattern to extract data
    # We can also use the parser to help us to extract the data, 
    # but here we present an alternative way to extract the data using regex.
    pattern = r"(\d+)\.\s+(.*?)\s+(\$[\d,]+\.\d+)\s+(\$[\d,]+\.\d+)\s+([\d.]+%)"

    # Use re.findall to extract data from the string
    matches = re.findall(pattern, item)

    # Create a DataFrame from the matches
    df = pd.DataFrame(matches, columns=['Rank', 'Player', 'Total_Year', 'Total_Overall', 'Percentage'])

    # Display the DataFrame
    df_total_earning[year] = df

    for top in tops:
        # URL of the page to scrape
        url = f'https://www.esportsearnings.com/history/{year}/top_players_{top}'

        # Send a GET request to the URL
        response = requests.get(url)

        # Check if the request was successful
        if response.status_code == 200:
            # Parse the content of the request with BeautifulSoup
            soup = BeautifulSoup(response.text, 'html.parser')
            # Find all divs with the class 'detail_list_box'
            detail_list_boxes = soup.find_all('div', class_='detail_list_box')
            item = detail_list_boxes[1].get_text(separator=' ', strip=True)
        else:
            print(f"Failed to retrieve data. Status code: {response.status_code}")

        item = item.split("% of Total ")[-1]

        # Define a regular expression pattern to extract data
        # We can also use the parser to help us to extract the data, 
        # but here we present an alternative way to extract the data using regex.
        pattern = r"(\d+)\.\s+(.*?)\s+(\$[\d,]+\.\d+)\s+(\$[\d,]+\.\d+)\s+([\d.]+%)"

        # Use re.findall to extract data from the string
        matches = re.findall(pattern, item)

        # Create a DataFrame from the matches
        df = pd.DataFrame(matches, columns=['Rank', 'Player', 'Total_Year', 'Total_Overall', 'Percentage'])
        # Append the data to the corresponding year's DataFrame
        df_total_earning[year] = pd.concat([df_total_earning[year], df], ignore_index=True)
```

# Top-1000 Highest Earning Players

```python
tops = [100,200,300,400,500,600,700,800,900,1000]

# URL of the page to scrape
url = 'https://www.esportsearnings.com/players/highest-earnings'

# Send a GET request to the URL
response = requests.get(url)

# List to hold all rows of data
all_rows_data = []

# Check if the request was successful
if response.status_code == 200:
    # Parse the content of the request with BeautifulSoup
    soup = BeautifulSoup(response.text, 'html.parser')

    # Find the tbody element
    tbody = soup.find('tbody')
    if not tbody:
        print("No tbody element found")
    else:
        # Find all 'tr' elements with class 'format_row highlight' within tbody
        rows = tbody.find_all('tr', class_='format_row highlight')
        
        # Extract and process data from each row
        for row in rows:
            # Extract the href attribute from the anchor tag in the second 'td' (player tag)
            # print(row)
            player_profile_link = row.find_all('td')[2].find('a', href=True)['href']
            player_profile_link = 'https://www.esportsearnings.com' + player_profile_link

            # Extracting data from each td element
            row_data = {
                'Rank': row.find_all('td')[0].get_text(strip=True),
                'Player Tag': row.find_all('td')[1].get_text(strip=True),
                'Player Name': row.find_all('td')[2].get_text(strip=True),
                'Total Earnings': row.find_all('td')[3].get_text(strip=True),
                'Primary Game': row.find_all('td')[4].get_text(strip=True),
                'Game Earnings': row.find_all('td')[5].get_text(strip=True),
                'Earnings Percentage': row.find_all('td')[6].get_text(strip=True),
                'Player Profile Link': player_profile_link  # Adding the player profile link
            }
            all_rows_data.append(row_data)
else:
    print(f"Failed to retrieve data. Status code: {response.status_code}")


for top in tops:
    # URL of the page to scrape
    url = f'https://www.esportsearnings.com/players/highest-earnings-top-{top}'

    # Send a GET request to the URL
    response = requests.get(url)

    # Check if the request was successful
    if response.status_code == 200:
        # Parse the content of the request with BeautifulSoup
        soup = BeautifulSoup(response.text, 'html.parser')

        # Find the tbody element
        tbody = soup.find('tbody')
        if not tbody:
            print("No tbody element found")
        else:
            # Find all 'tr' elements with class 'format_row highlight' within tbody
            rows = tbody.find_all('tr', class_='format_row highlight')

            # Extract and process data from each row
            for row in rows:
                # Extract the href attribute from the anchor tag in the second 'td' (player tag)
                player_profile_link = row.find_all('td')[2].find('a', href=True)['href']
                player_profile_link = 'https://www.esportsearnings.com' + player_profile_link

                # Extracting data from each td element
                row_data = {
                    'Rank': row.find_all('td')[0].get_text(strip=True),
                    'Player Tag': row.find_all('td')[1].get_text(strip=True),
                    'Player Name': row.find_all('td')[2].get_text(strip=True),
                    'Total Earnings': row.find_all('td')[3].get_text(strip=True),
                    'Primary Game': row.find_all('td')[4].get_text(strip=True),
                    'Game Earnings': row.find_all('td')[5].get_text(strip=True),
                    'Earnings Percentage': row.find_all('td')[6].get_text(strip=True),
                    'Player Profile Link': player_profile_link  # Adding the player profile link
                }
                all_rows_data.append(row_data)
    else:
        print(f"Failed to retrieve data. Status code: {response.status_code}")

# Create a DataFrame from the list of data
df = pd.DataFrame(all_rows_data)
```
